```python
def name_optimizer(row):
    row = row.replace('ё','е')
    row = row.replace('поселок ','')
    row = row.replace('городской ','')
    row = row.replace('деревня ','')
    row = row.replace('городского типа ','')
    row = row.replace('село ','')
    row = row.replace('городской ','')
    return row


df['locality_name'] = df['locality_name'].apply(name_optimizer)
```

```python
median_by_localion_list = df[(df['ceiling_height'] < 4) & (df['ceiling_height'] >= 2.5)].pivot_table(
    index = 'locality_name', values = 'ceiling_height', aggfunc = ['mean','median','count'])
median_by_localion_list.columns = ['mean_height','median_height', 'count']

display(median_by_localion_list.sort_values(by = 'median_height'))       
ax = median_by_localion_list['median_height'].hist(bins = 50, range = (2, 4), figsize = (12,6))
ax.set_xlabel("Высота потолков")
ax.set_ylabel("Количество объектов с реалистичными высотами")

median_by_localion_list.describe()
```
